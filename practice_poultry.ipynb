```
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```
data = pd.read_csv('data_chicken.csv')
data.head()
```

```
# here we will do some plotting
#fir we will plot how the weight varies with day
x = data.iloc[:,-7].values
y = data.iloc[:,-6].values
plt.scatter(x,y)
plt.xlabel('DAY')
plt.ylabel('WEIGHT')
plt.title('WOD')
plt.show()
```

```
# here we will do some plotting
#fir we will plot how the feed varies with day
x = data.iloc[:,-7].values
y = data.iloc[:,-5].values
plt.plot(x,y)
plt.xlabel('DAY')
plt.ylabel('feed')
plt.title('WOD')
plt.show()
```

```
# here we will do some plotting
#fir we will plot how the feed varies with weight
x = data.iloc[:,-5].values
y = data.iloc[:,-6].values
plt.plot(x,y)
plt.xlabel('feed')
plt.ylabel('weight')
plt.title('WOf')
plt.show()
```

```
# here we will do some plotting
#fir we will plot how the fcr varies with day
x = data.iloc[:,-7].values
y = data.iloc[:,-4].values
plt.plot(x,y)
plt.xlabel('day')
plt.ylabel('Fcr')
plt.title('WOf')
plt.show()
```

```
# here we will do some plotting
#fir we will plot how the profit varies with day
x = data.iloc[:,-7].values
y = data.iloc[:,-1].values
plt.plot(x,y)
plt.xlabel('day')
plt.ylabel('profit')
plt.title('profit')
plt.show()
```

```
# here we will do some plotting
#fir we will plot how the invested price varies with saleprice
x = data.iloc[:,-3].values
y = data.iloc[:,-2].values
plt.plot(x,y)
plt.xlabel('consumed')
plt.ylabel('sale')
plt.title('WOf')
plt.show()
```
